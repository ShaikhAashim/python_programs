```python
from bs4 import BeautifulSoup
import pandas as pd

website="local_website.html"
file=open(website,encoding="utf-8")
res=file.read()

tv_name = []
display_size = []
rating_star = []
rating_count = []
review_count = []
model_ID = []
launch_year = []
sound_quality = []
original_price = []
selling_price = []
discount = []

soup=BeautifulSoup(res,'html.parser')
#print(soup.prettify())

for i in soup.find_all('div',class_="RG5Slk")[0:20]:
    name = i.text.split('cm')[0].strip()
    tv_name.append(" ".join(name.split(' ')[:-1]))
    
    display=i.text.split('(')[1]
    display_size.append(display.split(')')[0])

for i in soup.find_all('div',class_="MKiFS6")[0:20]:
    rating_star.append(i.text)

for i in soup.find_all('span',class_="PvbNMB")[0:20]:
    rating_count.append(i.text.split('Ratings')[0])
    review=i.text.split('&')[1]
    #print(review.split(" ")[0])
    review_count.append(review.split(" ")[0])

for i in soup.find_all('div',class_="CMXw7N")[:20]:
    for j in i.find_all('li',class_="DTBslk"):
        if j.text.split(":")[0] == 'Model ID':
            model_ID.append(j.text.split(":")[1])

    for j in i.find_all('li',class_="DTBslk"):
        if j.text.split(":")[0] == 'Launch Year':
            launch_year.append(j.text.split(":")[1])

    for j in i.find_all('li',class_="DTBslk"):
        if j.text.split(":")[0] == 'Total Sound Output':
            sound_quality.append(j.text.split(":")[1])

for i in soup.find_all('div',class_='kRYCnD gxR4EY')[:20]:
    original_price.append(i.text)

for i in soup.find_all('div',class_='hZ3P6w DeU9vF')[:20]:
    selling_price.append(i.text)

for i in soup.find_all('div',class_="HQe8jr")[:20]:
    discount.append(i.text)

df = pd.DataFrame({
    'tv_name' : tv_name,
    'display_size' : display_size,
    'rating_star' : rating_star,
    'rating_count' : rating_count,
    'review_count' : review_count,
    'model_ID' : model_ID,
    'launch_year' : launch_year,
    'sound_quality' : sound_quality,
    'original_price' : original_price,
    'selling_price' : selling_price,
    'discount' : discount,
})
df
```
